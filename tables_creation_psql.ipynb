```python
##SQL TABLES

CREATE TABLE Analyst_Recommendations (
    id serial primary key,
    ticker VARCHAR(10) NOT NULL,
    date DATE NOT NULL,
    latest_stock_price FLOAT NOT NULL,
    high_target FLOAT NOT NULL,
    low_target FLOAT NOT NULL,
    mean_target FLOAT NOT NULL,
    median_target FLOAT NOT NULL,
    CONSTRAINT unique_ticker_date_target UNIQUE (ticker, date)
)

CREATE TABLE Analyst_Recommendations_Period(
    id SERIAL PRIMARY KEY,
    ticker VARCHAR(10) NOT NULL,
    date DATE NOT NULL,
    strong_buy INT NOT NULL,
    buy INT NOT NULL,
    hold INT NOT NULL,
    sell INT NOT NULL,
    strong_sell INT NOT NULL,
    CONSTRAINT unique_ticker_date_recommendations UNIQUE (ticker, date),
    CONSTRAINT fk_analyst_recommendations FOREIGN KEY (ticker, date)
        REFERENCES Analyst_Recommendations (ticker, date) ON DELETE CASCADE
)
```
